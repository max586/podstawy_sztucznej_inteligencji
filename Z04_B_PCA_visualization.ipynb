```python
samples = pd.read_csv("data/fish.csv", sep=";",decimal=",")
samples.head()
```

```python
sns.pairplot(data=samples, size=5)
plt.show()
```

```python
y = samples["Species"].values
X = samples.drop(["Species"],1).values
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.decomposition import PCA
```

## Odpowiedź: 3 wymiary są wystarczające

# Zad
Zwizualizuj dane w 2D

```python
model_3d = PCA(n_components=3)
samples_std = StandardScaler().fit_transform(samples)
data_3d = model_3d.fit_transform(samples_std)
df = pd.DataFrame(data_3d)
sns.pairplot(df)
plt.show()
```

```python
print(model_3d.explained_variance_)
print(model_3d.explained_variance_ratio_)
features = range(model_3d.n_components_)
plt.bar(features, model_3d.explained_variance_ratio_)
plt.xlabel('PCA feature')
plt.ylabel('variance')
plt.xticks(features)
plt.show()
```

```python
#calculate correlation
print('correlation 0-1: ',df[0].corr(df[1]))
print('correlation 1-2: ',df[1].corr(df[2]))
print('correlation 0-2: ',df[0].corr(df[2]))
```

```python
#reduce to 2 dimensions
pca_2 = PCA(n_components=2)
samples_std = StandardScaler().fit_transform(samples)
x_pca_3 = pca_2.fit_transform(samples_std)
df = pd.DataFrame(x_pca_3)
sns.pairplot(df)
plt.show()
```
